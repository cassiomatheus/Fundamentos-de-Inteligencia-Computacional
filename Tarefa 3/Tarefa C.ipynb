```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Função para desenhar o estado das moedas
def plot_coins(outcome, step):
    fig, ax = plt.subplots(figsize=(8,2))
    
    colors = {'CARA': 'gold', 'COROA': 'silver'}
    
    for idx, face in enumerate(outcome):
        circle = plt.Circle((idx*2, 1), 0.8, color=colors[face], ec='black')
        ax.add_artist(circle)
        ax.text(idx*2, 1, face[0], ha='center', va='center', fontsize=20, color='black')
    
    ax.set_xlim(-1, 9)
    ax.set_ylim(0, 2)
    ax.axis('off')
    plt.title(f"Passo {step}: Estado das Moedas")
    plt.show()

# 1º Passo: Definir o evento
outcome = ['CARA', 'CARA', 'CARA', 'COROA', 'CARA']
plot_coins(outcome, step=1)

# 2º Passo: Calcular a probabilidade de cara (p) e coroa (1-p)
p_cara = 0.5
p_coroa = 0.5

# Probabilidade desse evento específico (posição fixa)
# P = (0.5)^5
p_evento = (p_cara)**4 * (p_coroa)**1

print(f"Passo 2: Probabilidade do evento específico = {p_evento:.5f}")

# 3º Passo: Mostrar todos os possíveis resultados de 5 lançamentos
total_possibilidades = 2**5
print(f"Passo 3: Total de sequências possíveis = {total_possibilidades}")

# 4º Passo: Interpretar como teste binomial (4 caras, 1 coroa em ordem específica)
# O p-valor, nesse caso, seria a soma das probabilidades de obter um resultado 
# "tão extremo quanto" esse. Como cada sequência tem mesma probabilidade,
# e aqui é posição específica, o p-valor = probabilidade do evento.

print(f"Passo 4: O p-valor do evento é {p_evento:.5f}")

# 5º Passo: Visualizar
labels = ['Evento específico', 'Outro evento']
sizes = [p_evento, 1-p_evento]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.5f%%', startangle=90, colors=['lightblue', 'lightgray'])
ax.axis('equal')
plt.title('Passo 5: Proporção de Probabilidades')
plt.show()
```
