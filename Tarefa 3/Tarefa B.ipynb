```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, norm

# Parâmetros da distribuição Gama
forma = 2.0
escala = 2.0

# Tamanhos das amostras
tamanhos_amostras = [10, 30, 50, 100]

# Número de repetições
num_amostras = 1000

# Loop para cada tamanho de amostra
for n in tamanhos_amostras:
    medias = []

    # Gerar 1000 amostras e calcular a média de cada uma
    for _ in range(num_amostras):
        amostra = gamma.rvs(a=forma, scale=escala, size=n)
        medias.append(np.mean(amostra))

    # Mostrar o histograma das médias
    plt.hist(medias, bins=30, color='blue', alpha=0.7, density=True)

    # Desenhar a curva Normal teórica
    media_teorica = forma * escala
    desvio_padrao_teorico = (np.sqrt(forma) * escala) / np.sqrt(n)

    x = np.linspace(min(medias), max(medias), 100)
    plt.plot(x, norm.pdf(x, loc=media_teorica, scale=desvio_padrao_teorico), 'r-', lw=2)

    # Ajustes do gráfico
    plt.title(f"Tamanho da amostra: {n}")
    plt.xlabel("Médias das amostras")
    plt.ylabel("Densidade")
    #plt.grid(True)
    plt.show()
```
