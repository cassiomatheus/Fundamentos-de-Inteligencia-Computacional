#### Portanto o numero de amostras é 242485, a probabilidade de sucesso é 0.8 como é pedido na questão e o número de tentativas é 50.


```python
# Importação das bibliotecas
from scipy.stats import binom
from statistics import mean
from matplotlib import pyplot as plt
import numpy as np
from collections import Counter
```

```python
# Gerando dados seguindo uma distribuição binomial
nTG = 50  # número de tentativas
probG = 0.8  # probabilidade de sucesso ou "s"
TsizeG = 242485  # tamanhos de amostra
```

```python
np.random.seed(42) # Definindo a semente para reproduzir os resultados "Fixos"

# Gerando dados seguindo uma distribuição binomial
dados_binom = binom.rvs(n=nTG, p=probG, size=TsizeG)

print("Quatidade sucesso:",dados_binom[len(dados_binom)-1])

# Limites do eixo x
menor = min(dados_binom)
maior = max(dados_binom)

# Plotando o histograma dos dados
plt.hist(dados_binom, bins=25, density=True, alpha=0.6, color='orange')

# Plotando a função de massa de probabilidade
k = np.arange(menor, maior)
p = binom.pmf(k, n=nTG, p=probG)

plt.plot(k, p, 'k-', label=f'p={probG:.1f}, n={nTG}', linewidth=2)
plt.title('Distribuição Binomial')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.legend()
plt.show()
```

```python
# Valor do evento/amostra no ultimo elemento do array
x = dados_binom[len(dados_binom)-1]

#Calculando a probabilidade de acontecer "xq" vezes
binom.pmf(x, nTG, probG)

#Arredondando os valores para 8 casas decimais após a vírgula
round_off_values = np.around(binom.pmf(x, nTG, probG), decimals = 6)
round_off_values = round_off_values*100

#print(round_off_values)

print(f"A probabilidade de acontecer {dados_binom[len(dados_binom)-1]} (acertos/eventos) é de {round_off_values :.4f}%")
```

### Para encontrar o valor de \( p \), utilizou-se a fórmula da distribuição binomial, que permite calcular a probabilidade de se obter exatamente \( k \) sucessos em \( n \) tentativas independentes.

A fórmula da distribuição binomial é dada por:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

Onde:

1. \( P(X = k) \) é a probabilidade de obter exatamente \( k \) sucessos em \( n \) tentativas.
2. \( \binom{n}{k} \) é o coeficiente binomial, que pode ser calculado como \( \binom{n}{k} = \frac{n!}{k!(n - k)!} \).
3. \( p \) é a probabilidade de sucesso em cada tentativa.
4. \( (1 - p) \) é a probabilidade de falha em cada tentativa.
5. \( n \) é o número total de tentativas.
6. \( k \) é o número de sucessos desejados.


Dado:

n = 50<br>
k = 44 (Elemento 242485)<br>
p = 0.80<br>

Então:

$$
P(X = 44) = \frac{50!}{(50 - 44)! 44!} \times (0.80^{44}) \times (0.20^{50 - 44}) = 0.0553710  \quad \text{ou} \quad 5.5371\%
$$

Ou seja a propabilidade de acontecer 44 (acertos ou eventos) é de 5.5371%.


```python
# Comparando os resultados
val_cal = 5.5371
Dif = round_off_values - val_cal
print(f"Diferença entre {round_off_values} - {val_cal} = {Dif}")
```
