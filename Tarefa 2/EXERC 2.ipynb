#### 2.1 – O meu trabalho consiste na aplicação de modelos LLM para realizar a sumarização automática de documentos internos de uma empresa privada, com o objetivo de gerar novos documentos a partir das informações extraídas.

#### 2.2 – Portanto, a variável de interesse escolhida é o número de palavras (ou tokens) que cada documento possui, pois essa variável impacta diretamente no desempenho do processo de recuperação e geração de conteúdo pelas LLMs.

#### 2.3 – Considerando que o número de palavras por documento varia entre 2.000 e 8.000, adotei uma média de 5.000 palavras como valor de referência. A geração dos dados artificiais foi realizada com um desvio padrão de 500 palavras, simulando 300 amostras, que representam 300 documentos internos da empresa.

#### 2.4 – Com base nesses parâmetros, foi gerado um histograma com os dados simulados e uma curva de distribuição normal aproximada, utilizando a média e o desvio padrão definidos. Essa representação gráfica permite analisar o comportamento geral da variável em questão.


```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Dados artificiais
np.random.seed(42)

"""
loc=5000    - Média da distribuição (μ)
scale=100  - Desvio padrão da distribuição (σ)
size=300   - Quantidade de valores gerados(Amonto de dados)
""" 
tamanhos = np.random.normal(loc=5000, scale=500, size=360)  # tamanho em palavras

# Histograma
plt.hist(tamanhos, bins=20, density=True, alpha=0.6, color='blue')

# Curva normal ajustada
mu, sigma = np.mean(tamanhos), np.std(tamanhos)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', label=f'μ={mu:.1f}, σ={sigma:.1f}')
plt.title("Distribuição do Tamanho dos Documentos (Palavras)")
plt.xlabel("Número de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

### 📊 Estudo com Dados Simulados

Definimos um estudo no qual foram analisados 3 documentos com as seguintes quantidades de palavras:

| Documento | Nº de Palavras |
| --------- | -------------- |
| Doc A     | 7000           |
| Doc B     | 5100           |
| Doc C     | 3000           |

Sabendo que os dados simulados seguem uma distribuição com:

- Média (μ) = 5000 palavras
- Desvio padrão (σ) = 500
- Total de documentos simulados = 360

Podemos fazer uma primeira análise:

- **Doc A** está significativamente acima da média.
- **Doc B** está próximo da média, mas ainda acima.
- **Doc C** está abaixo da média.

---

### 📐 Padronização com Escore Z

O escore Z é calculado com a fórmula:

z-score = (Nº de Palavras do documento - média) / desvio padrão

Aplicando aos documentos:

- **z(Doc A)** = (7000 - 5000) / 500 = **4.0**
- **z(Doc B)** = (5100 - 5000) / 500 = **0.2**
- **z(Doc C)** = (3000 - 5000) / 500 = **-4.0**

---

### 🔍 Interpretação:

- 📄 **Doc A** está **4 desvios padrão acima** da média
- 📄 **Doc B** está **0.2 desvios padrão acima** da média
- 📄 **Doc C** está **4 desvios padrão abaixo** da média

---

### 📌 Conclusão:

O **Z-score** é uma métrica útil para entender a **posição relativa** de cada valor em relação à média da distribuição. Ele permite comparar, de forma padronizada, a quantidade de palavras nos documentos analisados, facilitando a identificação de valores atípicos ou padrões fora do comum.
