```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
```

```python
# Dados artificiais
np.random.seed(42)

"""
loc=5000    - Mdia da distribuio ()
scale=100  - Desvio padro da distribuio ()
size=300   - Quantidade de valores gerados(Amonto de dados)
"""
tamanhos = np.random.normal(loc=5000, scale=500, size=360)  # tamanho em palavras

# Histograma
plt.hist(tamanhos, bins=20, density=True, alpha=0.6, color='blue')

# Curva normal ajustada
mu, sigma = np.mean(tamanhos), np.std(tamanhos)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k-', label=f'={mu:.1f}, ={sigma:.1f}')
plt.title("Distribuio do Tamanho dos Documentos (Palavras)")
plt.xlabel("Nmero de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

```python
# Definindo parmetros da varivel
mu = 5000  # Mdia da distribuio
sigma = 100  # Desvio padro da distribuio
Size=2000 # Quantidade de valores gerados(Amostra de dados)

# Criando a populao artificial
np.random.seed(42)  # para reprodutibilidade
population = np.random.normal(loc=mu, scale=sigma, size=Size)

# Visualizando a populao
plt.figure(figsize=(10,6))
plt.hist(population, bins=30, density=True, alpha=0.6, color='blue')

# Curva normal ajustada
mu, sigma = np.mean(population), np.std(population)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k-', label=f'={mu:.1f}, ={sigma:.1f}')

plt.title("Distribuio do Tamanho dos Documentos (Palavras)")
plt.xlabel("Nmero de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

```python
# Tirando uma amostra
sample_size = 50
np.random.seed(42)  # para reprodutibilidade
sample = np.random.choice(population, size=sample_size)

# Curva normal ajustada
mu, sigma = np.mean(sample), np.std(sample)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Mostrando a sample distribution
plt.figure(figsize=(10,6))
#sns.histplot(sample, bins=10, kde=True, color='skyblue')
plt.hist(sample, bins=10, density=True, alpha=0.6, color='blue')
plt.plot(x, pdf, 'k-', label=f'={mu:.1f}, ={sigma:.1f}')
plt.title(f'Distribuio da Amostra (n={sample_size})')
plt.xlabel('Tempo (min)')
plt.ylabel('Frequncia')
plt.legend()
plt.show()
```

```python
# Criando sampling distribution
num_samples = 50
sample_means = []
np.random.seed(42)  # para reprodutibilidade
for _ in range(num_samples):
    sample_i = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample_i))

# Mostrando a sampling distribution
#plt.figure(figsize=(10,6))
#sns.histplot(sample_means, bins=20, kde=True, color='salmon')

######
# Curva normal ajustada
mu, sigma = np.mean(sample_means), np.std(sample_means)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Mostrando a sample distribution
plt.figure(figsize=(10,6))
#sns.histplot(sample, bins=10, kde=True, color='skyblue')
plt.hist(sample_means, bins=10, density=True, alpha=0.6, color='blue')
plt.plot(x, pdf, 'k-', label=f'={mu:.1f}, ={sigma:.1f}')
plt.title("Distribuio do Tamanho dos Documentos (Palavras)")
plt.xlabel("Nmero de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

### Tarefa B


##  Controle do Formato da Curva (Parmetro )

- **Quando  = 1**: A distribuio se torna **Exponencial**
- **Quando  < 1**: Curva com **pico no zero** e **cauda longa**
- **Quando  > 1**: Curva em forma de **sino assimtrico**
- **Quando   **: Aproxima-se de uma **distribuio Normal**

---

##  Controle da Disperso (Parmetro )

- **Valores maiores de **: **Esticam** a curva horizontalmente
- **Valores menores de **: **Comprimem** a distribuio

---

##  Caractersticas da Distribuio Gama(=2, =2)

- **Mdia terica**:  
  \[
  \mu = \alpha \times \theta = 2 \times 2 = 4
  \]

- **Varincia terica**:  
  \[
  \sigma^2 = \alpha \times \theta^2 = 2 \times 4 = 8
  \]

- **Assimetria**:  
  \[
  \text{Assimetria} = \frac{2}{\sqrt{\alpha}} = \frac{2}{\sqrt{2}} \approx 1.414
  \]  
  (Distribuio positivamente assimtrica)

---

##  Interpretao Prtica da Assimetria

- **Assimetria = 0**: Distribuio **simtrica** (como a Normal)
- **Assimetria > 0**: **Cauda alongada  direita** (valores altos mais provveis)
- **Assimetria < 0**: **Cauda alongada  esquerda** (valores baixos mais provveis)

**No caso com  = 2:**

- Assimetria  1.41  **Distribuio moderadamente assimtrica  direita**
- Formato tpico da distribuio Gama
- Explica por que as mdias amostrais tendem  Normal conforme \( n \to \infty \) (Teorema Central do Limite)

---

##  Frmula da Assimetria da Distribuio Gama

Para uma Gama com parmetros  (forma) e  (escala):

\[
\text{Assimetria} = \frac{2}{\sqrt{\alpha}}
\]
