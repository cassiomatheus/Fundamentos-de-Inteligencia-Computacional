```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
```

```python
# Dados artificiais
np.random.seed(42)

"""
loc=5000    - Média da distribuição (μ)
scale=100  - Desvio padrão da distribuição (σ)
size=300   - Quantidade de valores gerados(Amonto de dados)
""" 
tamanhos = np.random.normal(loc=5000, scale=500, size=360)  # tamanho em palavras

# Histograma
plt.hist(tamanhos, bins=20, density=True, alpha=0.6, color='blue')

# Curva normal ajustada
mu, sigma = np.mean(tamanhos), np.std(tamanhos)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k-', label=f'μ={mu:.1f}, σ={sigma:.1f}')
plt.title("Distribuição do Tamanho dos Documentos (Palavras)")
plt.xlabel("Número de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

```python
# Definindo parâmetros da variável
mu = 5000  # Média da distribuição
sigma = 100  # Desvio padrão da distribuição
Size=2000 # Quantidade de valores gerados(Amostra de dados)

# Criando a população artificial
np.random.seed(42)  # para reprodutibilidade
population = np.random.normal(loc=mu, scale=sigma, size=Size)

# Visualizando a população
plt.figure(figsize=(10,6))
plt.hist(population, bins=30, density=True, alpha=0.6, color='blue')

# Curva normal ajustada
mu, sigma = np.mean(population), np.std(population)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k-', label=f'μ={mu:.1f}, σ={sigma:.1f}')

plt.title("Distribuição do Tamanho dos Documentos (Palavras)")
plt.xlabel("Número de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

```python
# Tirando uma amostra
sample_size = 50
np.random.seed(42)  # para reprodutibilidade
sample = np.random.choice(population, size=sample_size)

# Curva normal ajustada
mu, sigma = np.mean(sample), np.std(sample)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Mostrando a sample distribution
plt.figure(figsize=(10,6))
#sns.histplot(sample, bins=10, kde=True, color='skyblue')
plt.hist(sample, bins=10, density=True, alpha=0.6, color='blue')
plt.plot(x, pdf, 'k-', label=f'μ={mu:.1f}, σ={sigma:.1f}')
plt.title(f'Distribuição da Amostra (n={sample_size})')
plt.xlabel('Tempo (min)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
# Criando sampling distribution
num_samples = 50
sample_means = []
np.random.seed(42)  # para reprodutibilidade
for _ in range(num_samples):
    sample_i = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample_i))

# Mostrando a sampling distribution
#plt.figure(figsize=(10,6))
#sns.histplot(sample_means, bins=20, kde=True, color='salmon')

######
# Curva normal ajustada
mu, sigma = np.mean(sample_means), np.std(sample_means)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Mostrando a sample distribution
plt.figure(figsize=(10,6))
#sns.histplot(sample, bins=10, kde=True, color='skyblue')
plt.hist(sample_means, bins=10, density=True, alpha=0.6, color='blue')
plt.plot(x, pdf, 'k-', label=f'μ={mu:.1f}, σ={sigma:.1f}')
plt.title("Distribuição do Tamanho dos Documentos (Palavras)")
plt.xlabel("Número de Palavras (Tokens)")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

```python
# Definindo parâmetros da variável
mu = 15  # média
sigma = 5  # desvio padrão

# Criando a população artificial
np.random.seed(42)  # para reprodutibilidade
population = np.random.normal(loc=mu, scale=sigma, size=100000)

# Visualizando a população
plt.figure(figsize=(10,6))
sns.histplot(population, bins=50, kde=True, color='lightgreen')
plt.title('População Artificial: Tempo de Atendimento')
plt.xlabel('Tempo (min)')
plt.ylabel('Frequência')
plt.show()

# Tirando uma amostra
sample_size = 50
sample = np.random.choice(population, size=sample_size)

# Mostrando a sample distribution
plt.figure(figsize=(10,6))
sns.histplot(sample, bins=15, kde=True, color='skyblue')
plt.title(f'Distribuição da Amostra (n={sample_size})')
plt.xlabel('Tempo (min)')
plt.ylabel('Frequência')
plt.show()

# Criando sampling distribution
num_samples = 1000
sample_means = []
for _ in range(num_samples):
    sample_i = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample_i))

# Mostrando a sampling distribution
plt.figure(figsize=(10,6))
sns.histplot(sample_means, bins=30, kde=True, color='salmon')
plt.title('Distribuição das Médias Amostrais')
plt.xlabel('Média do Tempo (min)')
plt.ylabel('Frequência')
plt.show()

# Avaliação de probabilidades
# Exemplo: Qual a probabilidade de o tempo médio de uma amostra ser menor que 14 minutos?
mean_sampling = np.mean(sample_means)
std_sampling = np.std(sample_means)

prob = norm.cdf(14, loc=mean_sampling, scale=std_sampling)
print(f"Probabilidade de a média da amostra ser menor que 14 minutos: {prob:.4f}")
```
